### Exploratory Analysis of Schizophrenia GWAS data

```python
import pandas as pd
```

```python
schizophrenia_df = pd.read_csv("daner_natgen_eas_eur_autosomes", sep='\t')
schizophrenia_df.sort_values('P',ascending= False)
schizo_df_filt= schizophrenia_df.loc[schizophrenia_df['P']<=5e-8]
schizo_df_filt.sort_values('OR',ascending= False)
```

```python
schizo_snps= pd.read_csv("schizophrenia_snps.txt", sep='\t', names=['SNP'])
schizo_df_final= schizo_df_filt.merge(schizo_snps, on="SNP")
schizo_df_final= schizo_df_final.sort_values('CHR',ascending=True)
schizo_df_final
```

### Confirming that Schizophrenia SNP's are in vcf files

### QQ Plots

```python
%pylab inline

## QQ Plots
fig = plt.figure()
ax = fig.add_subplot(111)

p_values = schizophrenia_df['P']

obs = (-np.log10(p_values)).sort_values(ascending=True)

# Generate some random data from uniform distribution
exp = -np.log10(np.random.uniform(0, 1, size=len(obs)))
exp.sort()

# Make a QQ plot
ax.scatter(exp, obs, s=5, color="blue")

#raise NotImplementedError()

ax.plot([0,10],[0,10], color="gray")
ax.set_xlabel("Expected -log10 P", size=15)
ax.set_ylabel("Observed -log10 P", size=15)
#ax.set_xlim(left=0, right=max(exp))
#ax.set_ylim(bottom=0, top=max(obs))
ax.set_xticklabels(ax.get_xticks(), size=12)
ax.set_yticklabels(ax.get_yticks(), size=12);
```
