## Analysis for Schizophrenia

```python
import pandas as pd
```

```python
schizophrenia_df = pd.read_csv("daner_natgen_eas_eur_autosomes", sep='\t')
schizophrenia_df.sort_values('P',ascending= False)
schizo_df_filt= schizophrenia_df.loc[schizophrenia_df['P']<=5e-8]
schizo_df_filt.sort_values('OR',ascending= False)
```

```python
schizo_snps= pd.read_csv("schizophrenia_snps.txt", sep='\t', names=['SNP'])
schizo_df_final= schizo_df_filt.merge(schizo_snps, on="SNP")
schizo_df_final= schizo_df_final.sort_values('CHR',ascending=True)
schizo_df_final
```

### Getting the final genotype for generated Offsprings

```python
%%bash

sed -i -e 's/0|1/1/g' -e 's/1|0/1/g' -e 's/0|0/0/g' -e 's/1|1/2/g' HG03108_HG01610_rand_1000_t.tsv
```

### Getting the Odds Ratio Data for Schizophrenia SNP's

```python
simulated_off = pd.read_csv("HG03108_HG01610_rand_1000_t.tsv", sep='\t')
simulated_off = simulated_off.rename(columns = {'ID': 'SNP'}, inplace = False)
simulated_off
```

```python
schizo_prs = schizo_df_final.merge(simulated_off, on='SNP')
schizo_prs
```

### Performing PRS Analysis for Schizophrenia

```python
column_names = schizo_prs.columns.to_list()
column_names = column_names[24:]
prs_individuals = []

for individual in column_names:
    schizo_prs['Score'] = schizo_prs.apply(lambda x: x[individual] * x['OR'], axis=1)
    score = schizo_prs['Score'].sum()
    list_item= [individual, score]
    prs_individuals.append(list_item)
    
```

```python
prs_ind_df = pd.DataFrame(prs_individuals, columns=['Individuals', 'Scores'])
```

```python
#prs_ind_df = prs_ind_df.rename(columns = {'0': 'Individuals', '1': 'Scores'}, inplace = False)
prs_ind_df
```

```python
# with open('final_prs_schizophrenia.txt', 'w') as filehandle:
#     for listitem in prs_individuals:
#         filehandle.write('%s\n' % listitem)

prs_ind_df.to_csv('final_prs_schizo_HG03108_HG01610_rand.csv', encoding='utf-8')
```
